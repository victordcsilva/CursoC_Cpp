### Primeiro Exemplo

- Stdlib.h é um arquivo cabeçalho da biblioteca de propósito geral padrão da linguagem de programação C. Ela possui funções envolvendo alocação de memória, controle de processos, conversões e outras. Ao finalizar o código com "system("pause");, o programa não será fechado no momento da execução. 

- void main ou int main: quer dizer que o programa irá retornar nada ou algum valor inteiro.

- A definição das variaveis se dá começando indicando o tipo dela. Ex: int a; char 'z'; float z;

- Para mostrar alguma mensagem na tela, se usa a função "printf();"

- Cada comando termina com o sinal de ponto e virgula (;)

- Para pular linha se usa o comando \n.

- A leitura de um valor de alguma variavel se dá pelo comando %d, %c, %s, %f, depende do tipo da variavel. 

```c++
#include<stdio.h>
#include <stdlib.h>

void main(){
    //definindo variavel
int a =5,b;
//imprimindo a variavel com concatenação
printf("\nO valor de a eh: %d", a);
    //pular linha
printf("\n");

    //mudificando valor
    a = 15;
    printf("\nO valor de a eh: %d", a);
    // lendo valores e escrevendo na memoria
    printf("\n");
    scanf("%d", &b);
    printf("\nO valor de b eh: %d", b);
    printf("\n");

//pausando
system("pause");

}
```

- Outra variavel é do tipo bool, indica se uma variavel é 1 (true) ou 0 (false).
- É possivel definir a quantidade de casas decimais após a virgula para números float, para duas casas se usa %.2f.
- Para trabalhar com variavel tipo bool é necessario a inclusão da biblioteca stdbool.h.

```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variaveis
    int a;
    float b;
    char c;
    bool d;

    //passando valores
    a = 5;
    b = 2.34;
    c = 'a';
    d = true; //true = 1 ou false = 0
    //escrevendo na tela
    printf("\n o valor de a = %d", a);
    printf("\n o valor de b = %.2f", b); //.2 mostra 2 casas depos da virgula
    printf("\n o valor de c = %c", c);
    printf("\n o valor de d = %d\n", d); //irá mostrar o valor 1
    //lendo valores
    scanf("%d", &a);
    scanf("%f", &b);
    scanf(" %c", &c); //é preciso colocar espaço antes das aspas duplas
    scanf("%d", &d);

    printf("\n o valor de a = %d", a);
    printf("\n o valor de b = %.2f\n", b);
    printf("\n o valor de c = %c\n", c);
    printf("\n o valor de d = %d\n", d);

    //pausando
    system("pause");

}
```

### Operações matematicas

- No c e c++ é possivel fazer as operações mais comuns (soma, subtração, divição e multiplicação), como outras como o resto da divisão, que retorna o resto de uma divisão entre dois numeros, e o valor absoluto de um número ou expressão matematica. 


```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variaveis
    int a = 6, b = 3;

    //soma
    printf("\nA soma de %d + %d = %d\n",a,b, a+b);
    //subtracao
    printf("\nA subtracao de %d - %d = %d\n",a,b, a-b);
    //divisao
    printf("\nA divisao de %d / %d = %d\n",a,b, a/b);
    //multiplicacao
    printf("\nA multiplicacao de %d * %d = %d\n",a,b, a*b);
    //resto da divisao
    printf("\nO resto da divisao %d + %d = %d\n",a,b, a%b);
    //valor absoluto
    printf("\nO valor absoluto de -3 eh = %d\n",abs(-3));
    //pausando
    printf("\n\n\ ");
    system("pause");

}
```

### Condicionais 

- Uma forma de testar se uma condição é verdadeira é utilizando o if.

- Pode ser simples ou composto, no caso de composto se usa o if, else if, else. 

- A comparação é testada com os seguintes códigos:
    - == (igual igual): Um resultado tem que ser igual ao outro.
    - != (diferente igual): Um resultado tem que ser diferente do outro.
    - < ou <= (menor ou menor igual): Um resultado tem que ser menor que o outro ou ser menor ou ter o mesmo valor.
    - > ou >=(maior ou maior igual): Um resultado tem que ser maior que o outro ou ser maior ou ter o mesmo valor. 

```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variaveis
    int a = 5,b = 10;

    //condicional
    if(a == 5){
        printf("\nA var a = 5");
    }
    //numero par ou impar
    if (b%2 == 1){
        printf("\nA var b eh impar");
    }
    else {
        printf("\nA var b eh par");
    }

    //condicional composta
    if (b == 3){
        printf("\nVar b eh igual a 3");
    }else if (b == 5) {
        printf("\nVar b eh igual a 5");
    }else{
        printf("\nVar b n eh igual a 3 ou 5");
    }

    //pausando
    printf("\n\n\ ");
    system("pause");

}
```

- Possivel testar para diferentes tipos de variaveis.

```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variavel
    int a = 5, b = 10, c = 15;
    char d = 'x';

    //maior
    if(a>2){
        printf("\n %d eh maior que 2",a);
    }
    //maior ou igual
    if (c>=b){
        printf("\n %d eh maior que %d", c,b);
    }
    //menor
    if (a<10){
        printf("\n %d eh menor que 10", a);
    }
    //diferente
    if (c!=10){
        printf("\n%d eh diferente de 10",c);
    }
    //difrente com char
    if (d!='a'){
        printf("\n %c eh diferente de a",d);
    }

    //pausando
    printf("\n\n ");
    system("pause");

}
```

### Conectivos Lógicos

- Muitas vezes é necessario testar mais de uma condição.
- Os conectivos lógicos são:
    - && (AND): As duas condições tem que ser verdadeira para o bloco ser aceito.
    - || (OR): Pelo menos uma das duas condições tem que ser verdadeira para o bloco ser aceito.

```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variavel
    int a = 10;

    //conectivo logico E (AND)
    //Se uma for falsa, não entra no bloco
    if (a > 5 && a <15 ){
        printf("\n A var a esta entre 5 e 15");
    }
    if (a > 5 && a >15 ){
        printf("\n A var a esta entre 5 e 15");
    }else{
        printf("\nAlguma condicional foi falsa");
        }
     //conectivo logico ou (OR)
    //Se uma for verdadeira, entra no bloco
    if (a > 5 || a >15 ){
        printf("\n A var a eh maior que 5 ou 15");
    }
    if (a > 5 || a >15 ){
        printf("\n Alguma condicional foi verdadeira");
    }else{
        printf("\nAs duas condicionais foram falsas");
        }


    //pausando
    printf("\n\n ");
    system("pause");

}
```

### Switch Case

- Funciona como um if, indica mais para ser usado como um sistema de escolhas.
- Possui a ação padrão, chamada de "default".

```c++
#include<stdio.h>
#include <stdlib.h>
#include <stdbool.h>

void main(){
    //definindo variavel
    int a = 5;
    char b = 'x';

    if(a == 1){
        printf("\nOpcao 1");
    }else if(a == 2){
        printf("\nOpcao 2");
    }else if(a == 3){
        printf("\nOpcao 3");
    }else{
        printf("\nOpcao invalida");
    }

    //switch case
    switch(a){
        case 1:
            printf("\nOpcao 1");
            break;
        case 2:
            printf("\nOpcao 2");
            break;
        case 3:
            printf("\nOpcao 3");
            break;
        default:
            printf("\nOpcao invalida");
            break;
        }

    switch(b){
    case 'x':
        printf("\nA letra eh x");
        break;
     default:
        printf("\nOpcao Invalida");
        break;
        }

    //pausando
    printf("\n\n ");
    system("pause");

}
```

### Função Rand (Número aleatório)

- Para gerar um número aleatorio em c/c++ é necessario pegar um tempo na memória. 
- Com algumas operações e escolhas de valores é possivel definir quais numeros aleatorios poderam ser sorteados.
- Necessario inclusão da biblioteca time.h.

```c++
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

void main(){
    //rand
    srand((unsigned)time(NULL)); //pega um tempo na memoria

    //var que recebe o resto da div do numero por 3
    int alea = rand()%3; //aleatorio entre 0, 1 e 2

    printf("\n%d",alea);

    //aleatorio entre 1 e 5
    int alea2 = (rand()%5)+1;
    printf("\n%d",alea2);


    //pausando
    printf("\n\n ");
    system("pause");

}
```

### Estrutura de repetição WHILE

- O while irá executar um comando enquanto uma condição se manter verdadeira;
- Necessario incrementar a condição, de forma que em algum momento ela não seja mais verdadeira e o while termine.
- Caso não acontece o programa entrará em "loop infinito", ou seja, irá continuar rodando eternamente. 

```c++
#include <stdio.h>
#include <stdlib.h>

void main(){
    //declarando variaveis
    int a = 1, b = 10;

    //contando até 10

    while(a <= 10){
        printf("\n%d",a);
        //incremento
        a = a + 1; //ou a++

    }
    //contagem regressiva

     while(b >= 1){
        printf("\n%d",b);
        //incremento
        b = b - 1; // ou b--

    }



    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

- O while pode ser escrito de outra forma, o do while.
- Enquanto o while irá executar comandos somente se alguma condição for verdadeira.
- O do while irá executar um comando primeiro e caso a condição continua verdadeira, ele irá repetir os comandos. 


```c++
#include <stdio.h>
#include <stdlib.h>

void main(){
    //declarando variaveis
    int a = 1, b = 10;

    //contando até 10

    //do while primeiro faz a operação, depois verifica a comparação
    //o while primeiro verifica, depois faz a operação
    do{
        printf("\n%d",a);
        //incremento
        a = a + 1; //ou a++

    }while(a <= 10);



    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

### Estrutura de repetição FOR

- O for funciona como um while, porém é mais rápido e escrito com menos comandos.
- Já é informado o tipo da variavel, o valor, a condição para repetição e o incremento.
- O incremento de 1 em 1 pode ser escrito como variavel++.

```c++
#include <stdio.h>
#include <stdlib.h>

void main(){
    //declarando variaveis
    int i;

    //contando até 10
    for(i=1;i<=10;i++){
        printf("\n%d",i);
    }
    //taboada
    printf("\nTabuada de 5: ");
    for(i=1;i<=10;i++){
        printf("\n 5 X %d = %d",i,5*i);
    }



    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

### Programa com sistema de escolhas

- O usuario irá passar um valor;
- Necessario ler o valor passado, para isso é usado a função "scanf();", é necessario escrever o simbolo & antes da variavel;

```c++
#include <stdio.h>
#include <stdlib.h>

void main(){
    //declarando variaveis
    int i;

    //confere e valida opcao
    while (i < 1 || i > 3  ){
    //interface de menu
        printf("\nEscolha uma opcao:");
        printf("\n1 - Opcao 1:");
        printf("\n2 - Opcao 2:");
        printf("\n3 - Opcao 3:\n");

    //lendo a opcao
        scanf("%d", &i);
    //resultado de acordo com a opcao
        switch(i){
            case 1:
                printf("\nOpcao 1");
                break;
            case 2:
                printf("\nOpcao 2");
                break;
            case 3:
                printf("\nOpcao 3");
                break;
            default:
                printf("\nOpcao errada");
                break;
        }}


    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

### Variavel fixa

```c++
#include <stdio.h>
#include <stdlib.h>
#define TAM 10 //criando variavel que não muda

void main(){
    //declarando variaveis
    printf("%d", TAM);

    //contagem

    for(int i=1;i<TAM;i++){
        printf("\n%d", i);
    }

    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

### Vetor

```c++
#include <stdio.h>
#include <stdlib.h>


void main(){
    //declarando variaveis
    int vetor[3];

    //passando valores para o vetor
    vetor[0] = 5;
    vetor[1] = 10;
    vetor[2] = 15;

    //Mostrando os valores

    printf("\nPosicao 0: %d", vetor[0]);
    printf("\nPosicao 1: %d", vetor[1]);
    printf("\nPosicao 2: %d", vetor[2]);

    for(int i=0;i<3;i++){
        printf("\nPosicao %d: %d",i,vetor[i]);
    }

    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

- Para escrever palavras, será necessario usar um vetor de char;
- No c se faz necessario limpar o buffer para não ocupar muita memoria 

```c++
#include <stdio.h>
#include <stdlib.h>


void main(){
    //declarando variaveis
    char palavra[225];

    //instrução
    printf("\nDigite uma palavra\n");

    //limpa o buffer
    setbuf(stdin,0);

    //Lê a string
    fgets(palavra,255,stdin);
    //limpa as casas não utilizadas

    palavra[strlen(palavra)-1]='\0';

    printf("%s",palavra);


    //pausando programa
    printf("\n\n" );
    system("pause");


}
```

- Já no c++ é mais facil, se usa string;
- É necessario a inclusão de duas bibliotecas, iostream e string;
- Antes do int main() (c++ não aceita void main()) se declara o comando "using namespace std;.
- Também não precisa mais de printf e scanf, trocados por cout e cin.


```c++
#include <stdio.h>
#include <stdlib.h>
#include <string>
#include <iostream>

using namespace std;//acessar as funcoes de string
int main(){

    //definindo string
    string palavra;

    //imprimindo
    cout << "\nDigite uma palavra:\n";
    //lendo
    cin >> palavra;
    //imprimindo
    cout << palavra;

    //concatenando

    cout<< "\nA palavra eh: " <<palavra;


    //pausando programa
    printf("\n\n" );
    system("pause");

    return 0;

}
```

### Matrizes

- Matrizes são vetores com mais de uma linha;

```c++
#include <stdio.h>
#include <stdlib.h>
#include <string>
#include <iostream>

using namespace std;
int main(){

    //definindo string
    int matriz[2][2];

    //passando valores

    matriz[0][0] = 1;
    matriz[0][1] = 2;
    matriz[1][0] = 3;
    matriz[1][1] = 4;

    for(int i=0;i<2;i++){
            for(int j=0;j<2;j++){
                //cout << "\ni = " << i <<", j = " <<j << endl; //mostra os indices
                cout << matriz[i][j] << " "; //mostra valores
            }
            cout << "\n";
    }


    //pausando programa
    printf("\n\n" );
    system("pause");

    return 0;

}
```
